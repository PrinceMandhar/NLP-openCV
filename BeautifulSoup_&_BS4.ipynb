```python
from textblob import TextBlob
from googlesearch import search
import requests
from bs4 import BeautifulSoup

# Get movie name input from the user
movie = input("Enter movie name: ")

# Create a search query
query = movie + ' user review'
print(f"Search Query: {query}")

# Function to get reviews from a URL
def get_reviews(url):
    reviews = []
    try:
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            # Assuming reviews are in <p> tags for simplicity
            for p in soup.find_all('p'):
                reviews.append(p.get_text())
    except Exception as e:
        print(f"Failed to fetch reviews from {url}: {e}")
    return reviews
    

# Search for the query and fetch URLs
urls = search(query, tld="co.in", num=10, stop=10, pause=2.0)

# Store the reviews in a list
all_reviews = []
for url in urls:
    print(f"Fetching reviews from: {url}")
    reviews = get_reviews(url)
    all_reviews.extend(reviews)

# Analyze the sentiment of each review
sentiments = defaultdict(int)
for review in all_reviews:
    blob = TextBlob(review)
    sentiment = blob.sentiment.polarity
    if sentiment > 0:
        sentiments['positive'] += 1
    elif sentiment < 0:
        sentiments['negative'] += 1
    else:
        sentiments['neutral'] += 1

# Print sentiment analysis results
print("Sentiment Analysis:")
print(f"Positive reviews: {sentiments['positive']}")
print(f"Negative reviews: {sentiments['negative']}")
print(f"Neutral reviews: {sentiments['neutral']}")

```
